```python
import tkinter as tk
from tkinter import messagebox
from PIL import Image, ImageTk

# Main window
root = tk.Tk()
root.title("Panaview Cinemas - Login")
root.geometry("800x600")
root.configure(bg="#5baaff")
root.resizable(False, False)

# GUI Elements, Top part of window
title = tk.Label(root, text="Panaview Cinemas", font=("Roboto", 24, "bold"), bg="#5baaff", fg="#ffffff")
title.pack(pady=50)

# Frame Container for login, Middle part of window
login_frame = tk.Frame(root, padx=40, pady=20, relief="solid",bg="#ffffff", highlightthickness=3, highlightbackground="#2141a1")
login_frame.place(relx=0.5, rely=0.5, anchor="center")

login_label = tk.Label(login_frame, text="Login", font=("Roboto", 20, "bold"),fg="#2260a3",bg="#ffffff")
login_label.grid(row=0, column=0, columnspan=2, pady=(0, 30))

create_frame = tk.Frame(root, padx=40, pady=20, relief="solid",bg="#5baaff")
create_frame.place(relx=0.5, rely=0.93, anchor="center")

create_text = tk.Frame(root, padx=0, pady=0, relief="solid",bg="#5baaff")
create_text.place(relx=0.5, rely=0.87, anchor="center")

# Username Label and Entry
username_label = tk.Label(login_frame, text="Username", font=("Roboto", 16, "bold"),fg="#2260a3",bg="#ffffff")
username_label.grid(row=1, column=0, sticky="w", pady=10)
username_entry = tk.Entry(login_frame, width=45, font=("Roboto", 14))
username_entry.grid(row=2, column=0, columnspan=2, pady=5)

# Password Label and Entry
password_label = tk.Label(login_frame, text="Password", font=("Roboto", 16, "bold"),fg="#2260a3",bg="#ffffff")
password_label.grid(row=3, column=0, sticky="w", pady=10)
password_entry = tk.Entry(login_frame, width=45, font=("Roboto", 14), show="*")
password_entry.grid(row=4, column=0, columnspan=2, pady=5)

# Bottom Part of window
create_label = tk.Label(create_text, text="Don't have an account?", font=("Roboto", 10, "bold"),fg="#ffffff",bg="#5baaff")
create_label.grid(row=0, column=0, columnspan=2, pady=0)



# Ticketer and App Logic (Static Data)
class Cinema:

    def create_account():
        global create
        create = tk.Tk()
        create.title("Panaview Cinemas - Create Account")
        create.geometry("800x600")
        create.configure(bg="#5baaff")
        create.resizable(False, False)


        # GUI Elements, Top part of window
        title = tk.Label(create, text="Panaview Cinemas", font=("Roboto", 24, "bold"), bg="#5baaff", fg="#ffffff")
        title.pack(pady=50)

        # Frame Container for create account, Middle part of window
        caccount_frame = tk.Frame(create, padx=40, pady=20, relief="solid",bg="#ffffff", highlightthickness=3, highlightbackground="#2141a1")
        caccount_frame.place(relx=0.5, rely=0.5, anchor="center")

        login_label = tk.Label(caccount_frame, text="Create Your Account", font=("Roboto", 20, "bold"),fg="#2260a3",bg="#ffffff")
        login_label.grid(row=0, column=0, columnspan=2, pady=(0, 30))

        # Username Label and Entry
        username_label = tk.Label(caccount_frame, text="Enter Your Username", font=("Roboto", 16, "bold"),fg="#2260a3",bg="#ffffff")
        username_label.grid(row=1, column=0, sticky="w", pady=10)
        username_entry = tk.Entry(caccount_frame, width=45, font=("Roboto", 14))
        username_entry.grid(row=2, column=0, columnspan=2, pady=5)

        # Password Label and Entry
        password_label = tk.Label(caccount_frame, text="Enter Your Password", font=("Roboto", 16, "bold"),fg="#2260a3",bg="#ffffff")
        password_label.grid(row=3, column=0, sticky="w", pady=10)
        password_entry = tk.Entry(caccount_frame, width=45, font=("Roboto", 14), show="*")
        password_entry.grid(row=4, column=0, columnspan=2, pady=5)

        create_frame = tk.Frame(create, padx=40, pady=20, relief="solid",bg="#5baaff")
        create_frame.place(relx=0.5, rely=0.93, anchor="center")

        sign_up = tk.Button(caccount_frame, text="Sign Up", font=("Roboto", 16, "bold"),bg="#5baaff", fg="#ffffff", width=20, command=Cinema.sign_up)
        sign_up.grid(row=5, column=0, columnspan=2, pady=30)

        # Bottom Part of window
        create_text = tk.Frame(create, padx=0, pady=0, relief="solid",bg="#5baaff")
        create_text.place(relx=0.5, rely=0.9, anchor="center")

        create_label = tk.Label(create_text, text="Have an account?", font=("Roboto", 10, "bold"),bg="#5baaff", fg="#ffffff")
        create_label.grid(row=0, column=0, columnspan=2, pady=0)

        signinc_account = tk.Button(create_text, text="Sign In", font=("Roboto", 16, "bold"),fg="#2260a3", bg="#ffffff", width=20, command=create.destroy)
        signinc_account.grid(row=5, column=0, columnspan=2)

    def sign_in():
        username = username_entry.get()
        password = password_entry.get()
        if username == "" and password == "":
            messagebox.showinfo("Login", "Login Successful!")
            Cinema.home_page()
        else:
            messagebox.showerror("Login", "Invalid username or password.")

    def sign_up():
        username = username_entry.get()
        password = password_entry.get()
        messagebox.showinfo("Login", "Account Successfully Created!")
        create.destroy()
            
    def home_page():
        global home
        home = tk.Toplevel()
        home.title("Panaview Cinemas")
        home.geometry("900x600")
        home.configure(bg="#5baaff")
        home.resizable(False, False)

        title = tk.Label(home, text="Now Showing", font=("Roboto", 24, "bold"), bg="#5baaff", fg="#ffffff")
        title.pack(pady=40)

        home_frame = tk.Frame(home, padx=40, pady=20, relief="solid", bg="#5baaff")
        home_frame.place(relx=0.5, rely=0.48, anchor="center")

        action_frame = tk.Frame(home, padx=90, pady=20, relief="solid", bg="#5baaff")
        action_frame.place(relx=0.5, rely=0.9, anchor="center")

        movie_names = ["Final Destination \nBloodlines", "Karate Kid: \nLegends", "Mission: Impossible - \nThe Final Reckoning", "Disney \nLilo & Stich"]
        movie_images = ["final_destinationpost.jpg", "karate_kidpost.jpg", "missionpost.jpg", "lilo_and_stitchpost.jpg"]
        Cinema.images = []
        for i in range(len(movie_names)):
            img = Image.open(movie_images[i])
            img = img.resize((197,290))
            tk_img = ImageTk.PhotoImage(img)
            Cinema.images.append(tk_img)
            btn = tk.Button(home_frame, text=movie_names[i], image=tk_img, compound="top", font=("Roboto", 15),fg="#2260a3", bg="#ffffff", command=lambda i=i: Cinema.movie_details(i))
            btn.pack(side="left", padx=10)

        exit_button = tk.Button(action_frame, text="Exit", font=("Roboto", 16, "bold"),fg="#2260a3", bg="#ffffff", width=20, command=Cinema.exit)
        exit_button.grid(row=5, column=1, padx=10)

        logout_button = tk.Button(action_frame, text="Log Out", font=("Roboto", 16, "bold"),fg="#2260a3", bg="#ffffff", width=20, command=Cinema.logout)
        logout_button.grid(row=5, column=0, padx=10)

    def movie_details(index):
        descriptions = [
            "A teenage girl has recurring nightmares of a tower \ncollapse in the 1960s She discovers that these nightmares \nare a premonition she inherited from her grandmother. \nThe grandmother predicted the collapse of the building \nand saved a group of people from death. \nDecades later, the granddaughter begins to have visions \nof her family members dying. She realizes that there is a \nsequence and must fight to prevent Death from \nreclaiming her family’s bloodline. \n\nGenre:        Horror, Now Showing, Slider \nRelease:     May 16, 2025 \nLanguage: English",
            "After kung fu prodigy Li Fong relocates to New York City, \nhe attracts unwanted attention from a local karate \nchampion and embarks on a journey to enter the ultimate \nkarate competition with the help of Mr. Han and Daniel \nLaRusso.\n\n\n\n\n\nGenre: Action, Drama, Family, Now Showing, Slider\nRelease: May 30, 2025\nLanguage: English",
            "Our lives are the sum of our choices. Tom Cruise is Ethan \nHunt in Mission: Impossible – The Final Reckoning.\n\n\n\n\n\n\n\n\nGenre: Thriller\nRelease: May 23, 2025\nLanguage: English",
            "A lonely Hawaiian girl befriends a runaway alien, helping \nto mend her fragmented family.\n\n\n\n\n\n\n\nGenre:  Action, Adventure, Comedy, Drama, Now Showing, \nSlider\nRelease: May 23, 2025\nLanguage: English"
        ]
        images = ["final_destinationpost.jpg", "karate_kidpost.jpg", "missionpost.jpg", "lilo_and_stitchpost.jpg"]
        titles = ["Final Destination Bloodlines", "Karate Kid: Legends", "Mission: Impossible - The Final Reckoning", "Disney Lilo & Stitch"]

        global detail
        detail = tk.Toplevel()
        detail.title("Panaview Cinemas")
        detail.geometry("900x600")
        detail.configure(bg="#5baaff")
        detail.resizable(False, False)

        tk.Label(detail, text=titles[index], font=("Roboto", 24, "bold"),bg="#5baaff",fg="#ffffff").pack(pady=30)

        img_frame = tk.Frame(detail, padx=0, pady=0, relief="solid", highlightthickness=3, highlightbackground="#2141a1" ,bg="#ffffff")
        img_frame.place(relx=0.2, rely=0.46, anchor="center")

        original = Image.open(images[index])
        resized = original.resize((275, 389))  
        img = ImageTk.PhotoImage(resized)

        img_label = tk.Label(img_frame, image=img)
        img_label.image = img
        img_label.pack()

        descrip_frame = tk.Frame(detail, padx=1, pady=10, relief="solid", bd=1, bg="#ffffff")
        descrip_frame.place(relx=0.675, rely=0.415, anchor="center")

        descrip_label = tk.Label(descrip_frame, text=descriptions[index], font=("Roboto", 14, "bold"),bg="#ffffff", justify="left", anchor="w")
        descrip_label.grid(row=1, column=0, sticky="w", pady=0)

        bknow_frame = tk.Frame(detail, padx=1, pady=10, relief="solid",bg="#5baaff")
        bknow_frame.place(relx=0.677, rely=0.755, anchor="center")

        bknow_button = tk.Button(bknow_frame, text="Time Slots", font=("Roboto", 16, "bold"),fg="#2260a3",bg="#ffffff", width=42, command=Cinema.show_times)
        bknow_button.grid(row=5, column=1)

        action_frame = tk.Frame(detail, padx=90, pady=20, relief="solid",bg="#5baaff")
        action_frame.place(relx=0.5, rely=0.9, anchor="center")

        tk.Button(action_frame, text="Exit", font=("Roboto", 16, "bold"),fg="#2260a3",bg="#ffffff", width=20, command=Cinema.exit).grid(row=5, column=1, padx=10)
        tk.Button(action_frame, text="Previous", font=("Roboto", 16, "bold"),fg="#2260a3",bg="#ffffff", width=20, command=detail.destroy).grid(row=5, column=0, padx=10)


    def show_times():
        global show
        show = tk.Toplevel()
        show.title("Panaview Cinemas")
        show.geometry("900x600")
        show.configure(bg="#5baaff")
        show.resizable(False, False)

        title = tk.Label(show, text="Show Times", font=("Roboto", 26, "bold"), bg="#5baaff", fg="#ffffff")
        title.pack(pady=10)

        time_frame = tk.Frame(show, padx=0, pady=0, relief="solid",bg="#5baaff")
        time_frame.place(relx=0.5, rely=0.48, anchor="center")

        time_button1 = tk.Button(time_frame, text="🕖7am - 8:50am      0/100 Seats💺", font=("Roboto", 30, "bold"),fg="#2260a3", bg="#ffffff", compound="left", width=35, command=Cinema.seat_arrangements)
        time_button1.grid(row=5, column=1, padx=10, pady=13)

        time_button2 = tk.Button(time_frame, text="🕘9am - 10:50am      0/100 Seats💺", font=("Roboto", 30, "bold"),fg="#2260a3", bg="#ffffff", compound="left", width=35, command=Cinema.seat_arrangements)
        time_button2.grid(row=6, column=1, padx=10, pady=13)

        time_button3 = tk.Button(time_frame, text="🕛12pm - 1:50pm      0/100 Seats💺", font=("Roboto", 30, "bold"),fg="#2260a3", bg="#ffffff", compound="left", width=35, command=Cinema.seat_arrangements)
        time_button3.grid(row=7, column=1, padx=10, pady=13)

        time_button4 = tk.Button(time_frame, text="🕒3pm - 4:50pm      0/100 Seats💺", font=("Roboto", 30, "bold"),fg="#2260a3", bg="#ffffff", compound="left", width=35, command=Cinema.seat_arrangements)
        time_button4.grid(row=8, column=1, padx=10, pady=13)


        action_frame = tk.Frame(show, padx=90, pady=20, relief="solid",bg="#5baaff")
        action_frame.place(relx=0.5, rely=0.9, anchor="center")

        exit_button = tk.Button(action_frame, text="Exit", font=("Roboto", 16, "bold"),fg="#2260a3",bg="#ffffff", width=20, command=Cinema.exit)
        exit_button.grid(row=5, column=1, padx=10)

        previous_button = tk.Button(action_frame, text="Previous", font=("Roboto", 16, "bold"),fg="#2260a3",bg="#ffffff", width=20, command=show.destroy)
        previous_button.grid(row=5, column=0, padx=10)

    def seat_arrangements():
        global seatwin
        seatwin= tk.Toplevel()
        seatwin.title("Panaview Cinemas")
        seatwin.geometry("900x600")
        seatwin.configure(bg="#5baaff")
        seatwin.resizable(False, False)

    def exit():
        root.destroy()
    def logout():
        home.destroy()
        detail.destroy()

sign_in = Cinema
sign_in = tk.Button(login_frame, text="Sign In", font=("Roboto", 16, "bold"),bg="#5baaff", fg="#ffffff", width=20, command=sign_in.sign_in)
sign_in.grid(row=5, column=0, columnspan=2, pady=30)

create_account = Cinema
create_account = tk.Button(create_frame, text="Create Account", font=("Roboto", 16, "bold"),fg="#2260a3", bg="#ffffff", width=20, command=lambda: (Cinema.create_account(), root.destroy))
create_account.grid(row=5, column=0, columnspan=2)

root.mainloop()
```
