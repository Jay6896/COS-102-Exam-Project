```python
import psycopg2
from psycopg2 import Error

def ensure_database_exists(db_name, user, password, host="localhost", port="5432", default_db="postgres"):
    """
    Connects to a default PostgreSQL database (e.g., 'postgres') to create
    the specified database if it does not already exist.

    Args:
        db_name (str): The name of the database to create.
        user (str): The database username (must have privileges to create databases).
        password (str): The database password.
        host (str, optional): The database host. Defaults to "localhost".
        port (str, optional): The database port. Defaults to "5432".
        default_db (str, optional): The initial database to connect to for creating
                                    the new database. Defaults to "postgres".
    """
    connection = None
    cursor = None
    try:
        # Attempt to connect to a default database (e.g., 'postgres')
        # This is necessary because you can't create a database while connected
        # to the database you are trying to create.
        connection = psycopg2.connect(
            database=default_db,
            user=user,
            password=password,
            host=host,
            port=port
        )
        connection.autocommit = True # Set autocommit for CREATE DATABASE statement
        cursor = connection.cursor()

        # SQL to create the database if it doesn't exist
        create_db_query = f"CREATE DATABASE {db_name};"
        cursor.execute(create_db_query)
        print(f"Database '{db_name}' created successfully or already exists.")

    except (Error) as error:
        print(f"Failed to create database '{db_name}': {error}")
        print("Please ensure your PostgreSQL user has privileges to create databases.")
    except Exception as e:
        print(f"An unexpected error occurred during database creation: {e}")
    finally:
        if cursor:
            cursor.close()
        if connection:
            connection.close()
            print(f"Connection to '{default_db}' closed.")


def create_tables(db_name, user, password, host="localhost", port="5432"):
    """
    Connects to a PostgreSQL database and creates 'users' and 'bookings' tables.

    Args:
        db_name (str): The name of the database where tables will be created.
        user (str): The database username.
        password (str): The database password.
        host (str, optional): The database host. Defaults to "localhost".
        port (str, optional): The database port.
    """
    connection = None
    cursor = None
    try:
        # Establish a connection to the PostgreSQL database (now that it should exist)
        connection = psycopg2.connect(
            database=db_name,
            user=user,
            password=password,
            host=host,
            port=port
        )
        cursor = connection.cursor()

        print(f"Connected to database '{db_name}' successfully!")

        # SQL statement to create the users table
        create_users_table_query = """
        CREATE TABLE IF NOT EXISTS users (
            id SERIAL PRIMARY KEY,
            username VARCHAR(50) UNIQUE NOT NULL,
            password VARCHAR(255) NOT NULL
        );
        """

        # SQL statement to create the bookings table
        create_bookings_table_query = """
        CREATE TABLE IF NOT EXISTS bookings (
            id SERIAL PRIMARY KEY,
            user_id INTEGER REFERENCES users(id) ON DELETE CASCADE,
            movie_name TEXT,
            time_slot TEXT,
            booked_seats TEXT,
            booking_time TIMESTAMP DEFAULT CURRENT_TIMESTAMP
        );
        """

        # Execute the CREATE TABLE statement for users
        cursor.execute(create_users_table_query)
        connection.commit()
        print("Table 'users' created successfully or already exists.")

        # Execute the CREATE TABLE statement for bookings
        cursor.execute(create_bookings_table_query)
        connection.commit()
        print("Table 'bookings' created successfully or already exists.")

    except (Error) as error:
        print(f"Failed to connect to '{db_name}' or create tables: {error}")
        print("Please check the following:")
        print("1. Is your PostgreSQL server running?")
        print("2. Are the database name, username, and password in the script correct?")
        print(f"3. Did the database '{db_name}' get created successfully?")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
    finally:
        if cursor:
            cursor.close()
        if connection:
            connection.close()
            print(f"PostgreSQL connection to '{db_name}' is closed.")

# --- How to use the functions ---
if __name__ == "__main__":
    # IMPORTANT: You MUST replace these placeholder values with your ACTUAL database credentials!
    # The user here should have privileges to create databases.
    DB_NAME = "panaview_db"   # The database you want to create and use
    DB_USER = "postgres"        # e.g., "postgres"
    DB_PASSWORD = "cos101"    # Your actual password
    DB_HOST = "localhost"            # Usually 'localhost' if running on the same machine
    DB_PORT = "5432"                 # Default PostgreSQL port

    # First, ensure the database itself exists
    ensure_database_exists(DB_NAME, DB_USER, DB_PASSWORD, DB_HOST, DB_PORT)

    # Then, create the tables within that database
    create_tables(DB_NAME, DB_USER, DB_PASSWORD, DB_HOST, DB_PORT)
```
